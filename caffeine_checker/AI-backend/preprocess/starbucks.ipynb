```python
# Convert string value to numeric value using pd.to_numeric function
menu['Caffeine (mg)'] = pd.to_numeric(menu['Caffeine (mg)'], errors='coerce')

# Get average value of menu['Caffeine (mg)']
menu['Caffeine (mg)'].mean()
```

```python
# Find NaN value and then fill it into mean value
menu.loc[menu['Caffeine (mg)'].isnull() == np.nan, 'Caffeine (mg)'] = 89.52054794520548
```

```python
menu['Caffeine (mg)']= menu['Caffeine (mg)'].fillna(menu['Caffeine (mg)'].mean())
```
